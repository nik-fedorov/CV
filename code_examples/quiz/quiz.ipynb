# Описание программы

Программа представляет из себя прохождение викторины на тему основ языка Python. Вопросы задаются по основным пройденным в курсе темам и имеют тестовую форму. Ответом на вопрос может служить как один, так и несколько вариантов ответа (обычно об этом сказано в условии вопроса). 

Программа начинает работу при запуске ячейки с кодом ниже. Взаимодествие пользователя с ней осуществляется через текстовую консоль (консольный интерфейс) под ячейкой с кодом. Чтобы выбирать пункты в меню или варианты ответа во время викторины, пользователю нужно вводит номера соответствующих пунктов/ответов в консоль и нажимать Enter. 

Изначально пользователю показывается главное меню программы. Здесь он может посмотреть подробные правила викторины, выйти из игры или начать викторину. Перед запуском викторины у пользователя спрашивают, в каком варианте он хочет проходить викторину. Возможные варианты:
- У вас есть только одна попытка, чтобы ответить на каждый вопрос. То есть в случае неверного ответа вам не будет предложено выбрать другой вариант ответа. В конце вам покажут, сколько всего было вопросов и сколько правильных ответов вы дали.
- Можно ошибаться неограниченное число раз. При вводе неправильного ответа программа всегда сообщает об этом и предлагает попробовать ответить на тот же вопрос снова. В конце программа показывает, сколько всего верных и неверных попыток ответить вы совершили за всю викторину

После выбора варианта квиза начинается сама викторина: пользователю задаются вопросы и предлагаются варианты ответов. При завершении викторины пользователь получает свои результаты и попадает в главное меню. Чтобы завершить викторину принудительно, пользователь в любой момент может ввести слово "quit" в качестве ответа на вопрос. Прохождение каждой викторины записывается в файл `sessions.txt`, при этом старые сессии не удаляются (новые данные *дописываются* в файл). 

```python
import datetime as dt


QUESTIONS_FILE_NAME = 'questions.txt'
SESSIONS_LOGGING_FILE_NAME = 'sessions.txt'

MAIN_MENU = '''ГЛАВНОЕ МЕНЮ
Для продолжения введите номер соответствующего пункта меню (например, если хотите начать викторину, введите 1 и нажмите Enter)
1. Начать викторину
2. Показать правила викторины
3. Выход'''

RULES = '''ПРАВИЛА ИГРЫ
Чтобы начать прохождение викторины, выберите 1 в главном меню.
Вам будет предложено два варианта викторины.
Первый: У вас есть только одна попытка, чтобы ответить на каждый вопрос. 
        То есть в случае неверного ответа вам не будет предложено выбрать другой вариант ответа. 
        В конце вам покажут, сколько всего было вопросов и сколько правильных ответов вы дали.
Второй: Можно ошибаться неограниченное число раз.
        При вводе неправильного ответа программа всегда сообщает об этом и предлагает попробовать ответить на тот же вопрос снова.
        В конце программа показывает, сколько всего верных и неверных попыток ответить вы совершили за всю викторину

Если вы хотите прервать викторину посреди игры, введите слово "quit" (без кавычек, в любом регистре). Вы попадете в главное меню. 

Во время квиза вам будут предлагаться вопросы с одним или несколькими вариантами ответа. Вводите свои варианты ответа по следующему правилу:
    *если вы хотите выбрать только один ответ, введите с клавиатуры ТОЛЬКО его цифру и нажмите Enter
    *если вы хотите ввести несколько ответов, то введите соответствующие им цифры через пробел в любом порядке и нажмите Enter
    *если вы хотите выйти из викторины, введите слово "quit" (без кавычек, в любом регистре) и нажмите Enter'''

QUIZ_OPTIONS_MENU = '''Выберите способ прохождения викторины:
1. Одна попытка. На каждый вопрос можно ответить только один раз
2. Можно ошибаться. Можете вводить ответ до тех пор, пока не ответите правильно
3. Назад в главное меню'''

GREETING = 'Здравствуйте! Добро пожаловать в программу-викторину, проверяющую ваше знание основ языка программирования Python!\n'
FAREWELL = 'Удачи на экзамене! До свидания!'
WRONG_INPUT_MSG_MAIN_MENU = '''Ваш ввод некорректен. Пожалуйста, введите ТОЛЬКО номер выбранного вами пункта меню!
Например, если вы хотите увидеть правила игры, введите с клавиатуры только цифру 2 (без точки) и нажмите Enter.'''
WRONG_INPUT_MSG_QUIZ_OPTIONS_MENU = '''Ваш ввод некорректен. Пожалуйста, введите ТОЛЬКО номер выбранного вами варианта викторины!
Например, если вы хотите иметь возможность ошибаться, введите с клавиатуры только цифру 2 (без точки) и нажмите Enter.'''
WRONG_INPUT_MSG_QUIZ = '''Ваш ввод некорректен. Пожалуйста, введите ТОЛЬКО номера выбранных вами вариантов ответа!
Например: *если вы хотите выбрать только ответ под номером 1, введите с клавиатуры ТОЛЬКО цифру 1 и нажмите Enter
          *если вы хотите ввести ответы с номерами 1, 2 и 3, то введите эти цифры через пробел в любом порядке и нажмите Enter
          *если вы хотите выйти из викторины, введите слово "quit" (без кавычек, в любом регистре) и нажмите Enter'''
WRONG_ANSWER_MSG = 'К сожалению, этот ответ неверный.'
RIGHT_ANSWER_MSG = 'Правильно!'
TRY_AGAIN = 'Попробуйте снова!'

EXCELLENT = 5 / 6
GOOD = 4 / 6
SATISFACTORY = 3 / 6


def show_main_menu():
    '''
    Отображает главное меню и проверяет ввод пользователя на корректность
    '''
    print(MAIN_MENU)
    ans = input().strip()
    while ans not in ['1', '2', '3']:  # пока ввод пользователя не станет корректным
        print(WRONG_INPUT_MSG_MAIN_MENU)
        ans = input().strip()  # снова считываем ввод пользователя
    return ans


def show_quiz_options():
    '''
    Отображает меню с выбором варианта квиза и проверяет ввод пользователя на корректность
    '''
    
    print(QUIZ_OPTIONS_MENU)
    ans = input().strip()
    while ans not in ['1', '2', '3']:  # пока ввод пользователя не станет корректным
        print(WRONG_INPUT_MSG_QUIZ_OPTIONS_MENU)
        ans = input().strip()  # снова считываем ввод пользователя
    return ans


def get_user_answer(n_options):
    '''
    Функция печатает приглашение к вводу пользователем ответа на очередной вопрос викторины.
    После чего проверяет его на корректность и, когда ввод оказывается корректным, возвращает его
    '''
    user_ans = input('Ваш ответ: ').lower().strip()  # печатаем приглашение к вводу, приводим ответ к нижнему регистру
    
    # цикл ниже работает до тех пор, пока пользователь не введет слово "quit" или 
    # последовательность чисел, разделенных пробелом, от 1 до количества предложенных ответов
    while user_ans != 'quit' and not all(map(lambda x: x.isdigit() and 0 < int(x) <= n_options, user_ans.split())):
        print(WRONG_INPUT_MSG_QUIZ)
        user_ans = input('Ваш ответ: ').lower().strip()
    return user_ans


def get_quiz_summary(quiz_type, n_right, n_wrong):
    '''
    Формирует и возвращает строку с результатами викторины в зависимости от ее варианта.
    Результаты включают в себя: количество правильных и неправильных ответов, 
    процент правильных ответов, итоговая оценка (отлично, хорошо, удовлетворительно или плохо)
    '''
    summary = 'Итог:\n'
    if quiz_type == '1':  # Одна попытка
        summary += 'Отвечено верно на ' + str(n_right) + ' вопросов из ' + str(n_right + n_wrong)
    else:  # Можно ошибаться
        summary += 'Всего дано ' + str(n_right + n_wrong) + ' ответов.\n'
        summary += 'Из них ' + str(n_right) + ' верных и ' + str(n_wrong) + ' неверных'
    share = n_right / (n_right + n_wrong)  # доля правильных ответов
    summary += '\nПроцент правильных ответов: ' + str(share * 100) + '\nВаш результат: '
    if share >= EXCELLENT:
        summary += 'Отлично'
    elif share >= GOOD:
        summary += 'Хорошо'
    elif share >= SATISFACTORY:
        summary += 'Удовлетворительно'
    else:
        summary += 'Вам следует еще подучить теорию перед экзаменом!'
    return summary


def read_questions_from_file(filename):
    '''
    Читает вопросы, варианты ответа и правильные ответы из файла filename. 
    После чего записывают вс. эту информацию в словарь data, в котором ключом 
    является вопрос, а значением - список, в котором 
    * первый элемент - список с вариантами ответа
    * второй элемент - список правильных ответов
    * все последующие элементы - строки-вводы пользователя (его попытки ответить на вопрос)
    Вводы пользователя добавляются позже во время игры.
    Возвращает словарь data
    '''
    with open(filename, 'r', encoding='utf-8') as f:  # открываем файл с вопросами на чтение
        lines = list(map(lambda x: x.strip(), f.readlines()))  # читаем строки файла, убираем \n в конце строк
    
    # каждый вопрос занимает 3 строки: на первой сам вопрос, на второй варианты ответа, разделенные символом |
    # на третьей правильный ответ (если правильный ответ включает несколько пунктов, их номера разделены |)
    questions = lines[0::3]
    questions_options = list(map(lambda x: x.split('|'), lines[1::3]))
    right_answers = list(map(lambda x: x.split('|'), lines[2::3]))
    
    data = {}
    for q, q_op, q_right in zip(questions, questions_options, right_answers):
        data[q] = [q_op, q_right]
    return data


def write_session_log_in_file(filename, data, session_start_dt, quiz_type, n_right, n_wrong):
    '''
    Записывает информацию об очередной викторине в файл filename. Эта информация включает:
    * дата и время начала викторины
    * вопросы, задаваемые программой
    * все ответы (попытки) пользователя
    * правильные ответы
    * результат викторины
    '''
    with open(filename, 'a', encoding='utf-8') as f:  # открываем файл на дописывание
        f.write('\n\n\nНачало викторины: ' + str(session_start_dt) + '\n')
        if quiz_type == '1':  # Одна попытка
            quiz_option = 'Одна попытка'
        else:  # Можно ошибаться
            quiz_option = 'Можно ошибаться'
        f.write('Вариант викторины: ' + quiz_option + '\n')
        for q in data:
            options, right_ans, user_ans = data[q][0], data[q][1], data[q][2:]
            if not user_ans:  # на этот и следующие вопросы пользователь не дал ответ,
                break         # потому что закончил викторину принудительно (ввел "quit")
            f.write('\nВопрос: ' + q + '\n')
            f.write('Варианты ответа:\n' + '\n'.join(options) + '\n')
            f.write('Ответы пользователя:\n' + '\n'.join(user_ans) + '\n')
            f.write('Правильный ответ: ' + ' '.join(right_ans) + '\n')
        f.write('\n' + get_quiz_summary(quiz_type, n_right, n_wrong) + '\n')


def main():
    '''
    Основная функция, содержащая основной цикл работы программы. 
    Осуществляет перемещение между меню, чтение вопросов из файла, 
    проведение викторины, запись очередной викторины в файл, 
    здоровается и прощается с пользователем
    '''
    print(GREETING)
    while True:  # основной цикл программы, показывающий главное меню
        user_ans = show_main_menu()  # показываем главное меню и получаем ответ пользователя
        if user_ans == '3':  # выход
            print(FAREWELL)
            break  # выход из основного цикла
        elif user_ans == '2':  # напечатать правила
            print(RULES + '\n')
        else:  # начать викторину
            quiz_type = show_quiz_options()  # показываем меню с вариантами квиза и получаем ответ пользователя
            if quiz_type != '3':  # идем дальше только если пользователь НЕ выбрал "Выход"
                session_start_dt = dt.datetime.now()  # запоминаем дату и время начала викторины, чтобы потом записать в историю
                data = read_questions_from_file(QUESTIONS_FILE_NAME)
                n_right, n_wrong = 0, 0  # инициализируем счетчики правильных и неправильных ответов

                for q in data:
                    q_op, right_ans = data[q]
                    print(q + '\n' + '\n'.join(q_op))
                    user_ans = get_user_answer(len(q_op))  # получаем ответ пользователя на очередной вопрос квиза
                    data[q].append(user_ans)  # запоминаем ответ в словаре data
                    if quiz_type == '1':  # Одна попытка
                        if user_ans == 'quit':
                            break
                        elif set(user_ans.split()) == set(right_ans):  # проверка на правильность ответа
                            print(RIGHT_ANSWER_MSG)
                            n_right += 1
                        else:
                            print(WRONG_ANSWER_MSG)
                            n_wrong += 1
                    elif quiz_type == '2':  # Можно ошибаться
                        # пока пользователь не выйдет или не ответит правильно
                        while user_ans != 'quit' and set(user_ans.split()) != set(right_ans):
                            print(WRONG_ANSWER_MSG, TRY_AGAIN)
                            n_wrong += 1
                            user_ans = get_user_answer(len(q_op))
                            data[q].append(user_ans)
                        if user_ans == 'quit':
                            break
                        else:
                            print(RIGHT_ANSWER_MSG)
                            n_right += 1
                
                write_session_log_in_file(SESSIONS_LOGGING_FILE_NAME, data, session_start_dt, 
                                          quiz_type, n_right, n_wrong)
                print('\n' + get_quiz_summary(quiz_type, n_right, n_wrong) + '\n\n')


main()
```
